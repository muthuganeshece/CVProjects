```python
import numpy as np
import pandas as pd
```

```python
file_path = r"F:\Work\Misc\Mentees\SodiumDisposal\combined_data_grouped_with_padding_flag_Modified.xlsx"
outpath = "/".join(file_path.split("\\")[:-1])
raw_data = pd.read_excel(file_path, sheet_name='combined_data_grouped_with_padd')
```

```python
raw_data.head()
```

# Training, Test, Validation data generation

```python
ignore_cols = ['full path', 'Train', 'Test', 'Val']
```

```python
for i in ['Train', 'Test', 'Val']:
    train_data = raw_data[raw_data[i] == 1]
    unique_paths = train_data['full path'].unique()
    subset = pd.DataFrame()
    for path in unique_paths:
        subsample = train_data[train_data['full path'] == path]
        start_index = min(list(subsample.index))
        end_index = max(list(subsample.index)) + 2
        subsample = raw_data.iloc[start_index:end_index, :]
        if subset.empty:
            subset = subsample
        else:
            subset = pd.concat([subset, subsample])
    subset.drop(ignore_cols, axis=1).to_csv(outpath + "/" + i.lower() + '_set.csv', index=False)
```
